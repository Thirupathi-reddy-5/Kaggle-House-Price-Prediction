# Filling the null Values in House Price Dataset

Importing the necessary libraries

```python
import pandas as pd
import numpy as np
```

Reading the dataset

```python
df=pd.read_csv("/content/Kaggle House Price.csv")
```

Filtering all the features having  null values into a list

```python
features_with_null=[feature for feature in df.columns if df[feature].isnull().sum()>1]
```

```python
features_with_null
```

Finding the percentage of nulll values in each feature

```python
for feature in features_with_null:
  print(feature,np.round(df[feature].isnull().mean()*100,4))
```

Filtering out the categoricalk features having null values into a new list

```python
categorical_variables_null=[feature for feature in features_with_null if df[feature].dtypes=='O']
```

```python
categorical_variables_null
```

Creating a new dataset copy into the dataframe  "data" to fill the null values

Filling all the null values in the categorical features with a new feature "missing"

```python
data=df.copy()
for feature in categorical_variables_null:
  data[feature]=data[feature].fillna('Missing')
```

Checking the percentage of null values after filling

```python
for feature in categorical_variables_null:
  print(feature,np.round(data[feature].isnull().mean()*100,2))
```

#filtering the  numerical features having null values into a new array

```python
Numeric_values_null=[feature for feature in features_with_null if feature not in categorical_variables_null]

```

```python
Numeric_values_null
```

Top 5 rows of  features in numeric values

```python
df[Numeric_values_null].head()
```

```python
for feature in Numeric_values_null:
  print(feature,np.round(data[feature].isnull().mean()*100,2))
```

# filling the null values in numerical features

Creating new features to represent the null values with 1 and reamining with 0

Filling the null values with median

```python
for feature in ['LotFrontage','MasVnrArea']:
  data[feature+' NAn']=np.where(data[feature].isnull(),1,0)
  data[feature]=data[feature].fillna(data[feature].median())
```

```python
for feature in Numeric_values_null:
  print(feature,np.round(data[feature].isnull().mean()*100,2))
```

# Handling Numerical values

# Applying logarithmic transformation for numerical variables which doesn't contain zero


here we apply logarithmic transformation for only continuous numeric features because there is no use of applying it for discrete and categorical numeric variables

Discrete include no.of children , etc applying log transformation for this doesn't make any sense

In the same way categorical N.V like Zip code also doesn't make any sense

```python
data.shape
```

```python
Numerical_values=[feature for feature in data.columns if data[feature].dtypes!='O' and len(data[feature].unique())>20]
```

```python
Numerical_values_for_log=[feature for feature in Numerical_values if 0 not in data[feature].unique()]
```

```python
Numerical_values_for_log
```

```python
for feature in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']:
  data[feature]=np.log(data[feature])
```

# Now  we have to handle the categorical variables

#  Handling the rare categorical variables

```python
categorical_features=[feature for feature in data.columns if data[feature].dtypes=='O']
```

```python
len(data['1stFlrSF'])
```

Finding the percentage of each category in categorical features

```python
for feature in categorical_features:
  print(feature,np.round(data[feature].value_counts()/len(data[feature])*100,2))
```

```python
pd.set_option('display.max_rows',None)

for feature in categorical_features:
  features=data[feature].value_counts(normalize=True)
  rare_features=features[features<0.01].index
  data[feature]=np.where(data[feature].isin(rare_features),'Rare_var',data[feature])

```

```python
data.head(10)
```

# Handling the remaining categorical variables

```python
for feature in categorical_features:
  print(feature,data[feature].nunique())
```

#applying the target encoding based on the mean sale price of each category

```python
for feature in categorical_features:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
data.head(10)
```

Now we have handled all the categorical varaibles and left with numerical variables

So to handle these numeric variables we are going to apply scaling using  min max scaler

```python
feature_scale=[feature for feature in data.columns if feature not in ['Id','SalePrice']]
len(feature_scale)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(data[feature_scale])
```

```python
data=pd.concat([data[['Id','SalePrice']].reset_index(drop=True),pd.DataFrame(scaler.transform(data[feature_scale]),columns=feature_scale)],axis=1)
```
